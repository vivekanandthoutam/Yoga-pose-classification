```python
pairs_arr=[[0,2],[2,4],[4,6],[0,8],[8,10],[10,12],[0,14],[14,16],[16,18],[0,20],[20,22],[22,24]]
pairs_arr_cn=["[0,2]","[0,2]","[2,4]","[2,4]","[4,6]","[4,6]","[0,8]","[0,8]","[8,10]","[8,10]","[10,12]","[10,12]","[0,14]","[0,14]","[14,16]","[14,16]","[16,18]","[16,18]","[0,20]","[0,20]","[20,22]","[20,22]","[22,24]","[22,24]"]
```

```python
np.arange(11)
```

```python
df_pv=pd.DataFrame(0, index=np.arange(df.shape[0]), columns=pairs_arr_cn)
```

```python
df_pv
```

```python
prac_df_pv=df_pv
```

```python
for i in range(len(pairs_arr)):
  a=pairs_arr[i][0]
  b=pairs_arr[i][1]
  for j in range(df.shape[0]):
    prac_df_pv.loc[j][i*2]=df.loc[j][b]-df.loc[j][a]
    prac_df_pv.loc[j][i*2+1]=-1*(df.loc[j][b+1]-df.loc[j][a+1])
```

```python
prac_df_pv
```

```python
prac_df_pv.columns
```

```python
pairs_arr_cng=["[0,2]","[2,4]","[4,6]","[0,8]","[8,10]","[10,12]","[0,14]","[14,16]","[16,18]","[0,20]","[20,22]","[22,24]"]
```

```python
df_gv=pd.DataFrame(0, index=np.arange(df.shape[0]), columns=pairs_arr_cng)
```

```python
prac_df_gv=df_gv
```

```python
prac_df_gv
```

```python
for i in range(len(pairs_arr)):
  for j in range(df.shape[0]):
    if (prac_df_pv.loc[j][i*2]!=0 or prac_df_pv.loc[j][i*2+1]!=0):
       val=(prac_df_pv.loc[j][i*2])/(((prac_df_pv.loc[j][i*2]**2)+(prac_df_pv.loc[j][i*2+1]**2))**(0.5))
       if ( (prac_df_pv.loc[j][i*2+1]>0)):
           prac_df_gv.loc[j][i]=math.acos(val)*180*(1/3.141592653589793)
       else:
           prac_df_gv.loc[j][i]=360 - math.acos(val)*180*(1/3.141592653589793)
    else:
           prac_df_gv.loc[j][i]=1000
           
    
```

```python
prac_df_gv
```

```python
prac_df_gv.to_csv('/content/sample_data/angles_Sathak_shav.csv')
```
